```python
# 리스트 -> 리스트: 컴프리헨션

# 계산에서 사용되는 계산 데이터셋(원본X)
# 판매액에서 정수가 아닌 데이터 빼고 만들어주기
values = [r for r in sales if isinstance(r["판매액"], int)]
len(values)
```

```python
# 1. 최댓값/최솟값 찾기
# 판매액이 가장 큰 거래와 가장 작은 거래를 찾으시오.
max(values, key= lambda x: x["판매액"])
```

```python
min(values, key= lambda x: x["판매액"])
```

```python
# 2. Top-N 추출
# 판매액이 가장 큰 거래 Top3를 출력하시오.
# 정렬 / 람다 사용

sort_values = sorted(values, key= lambda x: x["판매액"], reverse=True)

rank_num = 1
for r in sort_values[:3]:
    print(f"{rank_num}등 {r}")
    rank_num += 1
```

```python
# 3. 빈도수 계산
# 상품별 판매 건수를 구하시오.

count = {} #dict 형태로 결과값 집어넣는 곳
for r in sales:
    key_name = r["상품"]
    if key_name not in count:
        count[key_name] = 0
    count[key_name] = count[key_name] + 1

print(count)

for k, v in count.items():
    print(f"{k} : {v}건")
```

```python
# 4. 그룹별 집계(Group By)
# 지역별 총 판매액을 구하시오.

local_sum = {}

for r in sales:
    price = r["판매액"]
    #price "", "결측"이면
    if not isinstance(price, int):
        continue #다음 반복으로 건너뛰기

    key_name = r["지역"]

    if key_name not in local_sum:
        local_sum[key_name] = 0
    
    local_sum[key_name] += price

for l, p in local_sum.items():
    print(f"{l} : {p}")
```

```python
# 5. 중복 제거
# 판매된 지역 목록만 출력하시오. (set사용)

regions = set()
for r in sales:
    regions.add(r["지역"])

print(regions)
```

```python
# 7. 순위 계산(Ranking)
# 판매액 기준 순위를 구하시오.
# 2번 문제에서 사용한 sort_values

rank = 1
for r in sort_values:
    print(f"{rank}위 - {r}")
    rank += 1
```

```python
# 8. 이동평균(Moving Average)
# 판매액 / 슬라이딩 윈도우

# 슬라이딩 윈도우에 사용될 리스트
prices = []

for r in sales:
    if isinstance(r["판매액"], int):
        prices.append(r["판매액"])

# prices

window = 3
for i in range(len(prices) - (window - 1)):
    total = sum(prices[i:i+window])
    avg = round(total / window, 3)
    print(f"{i+1}~{i+window}일 평균 판매액 : {avg}원")
```

```python
# 9. 누적평균(Running Average)
prices = []
for r in sales:
    if isinstance(r["판매액"], int):
        prices.append(r["판매액"])

prefix = [0]

for p in prices:
    prefix.append(prefix[-1] + p)

avgs = []
for i in range(1, len(prefix)):
    avgs.append(prefix[i] / i)

avgs
```

```python
# 10. 이상치 탐지(Threshold)
# 9999999를 찾아라. or 100만원 이상을 이상치로 판단.

for r in sales:
    if isinstance(r["판매액"], int):
        if r["판매액"] >= 2000000:
            print("이상치 발견", r)
```

```python
# 11. 결측치 처리
# "","결측"을 평균이나 중앙값으로 변경.
# try 예외처리 또는 if 조건분기

#평균
avg = sum(prices) / len(prices)

#중앙값
#print(len(prices)) #짝수

prices.sort()
print(prices)

median = 0 # 중앙값 초기화

if len(prices) % 2 == 1: #홀수이면
    median = prices[len(prices) // 2]
    # 중앙값 1개 --> [0, 1, 2, 3, 4] 
    # --> 마지막 인덱스 나누기 2의 몫(4//2)을 인덱스로 or 사이즈 나누기 2의 몫(5//2)
else: # 짝수이면
    median = (prices[len(prices) // 2] + prices[len(prices) // 2 - 1]) / 2
    # 중앙값 2개 / 2 [0, 1, 2, 3, 4, 5]
    #마지막 인덱스 2의 몫(5//2)과 마지막 인덱스 2의 몫+1(5//2 + 1)
    # 사이즈 2의 몫(6//2)과 사이즈 2의 몫-1 (6//2 - 1)

print("원본", median)

for r in sales:
    if not isinstance(r["판매액"], int):
        r["판매액"] = median

print("결측치 → 중앙값", sales)
```

```python
# 12. 범주별 통계

result = {}

for r in sales:
    p_name = r["상품"]
    price = r["판매액"]
    total_cnt = r["수량"]
    if p_name not in result:
            result[p_name] = {"합계":0, "건수":0, "최대":price, "최소":price, "총 판매량":0}
            # {
            # "키보드" : {"합계":0, "건수":0, "최대":0, "최소":0, "총 판매량":0}
            # "모니터" : {"합계":0, "건수":0, "최대":0, "최소":0, "총 판매량":0}
            # ...
            # }
    info = result[p_name]
    info["합계"] += price
    info["건수"] += 1
    info["총 판매량"] += total_cnt

    #최대/최소
    #result["노트북"]
    if info["최대"] < price:
          info["최대"] = price
    if info["최소"] > price:
          info["최소"] = price

print(result)
```
