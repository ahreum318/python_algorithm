```python
# 1. 아래와 같이 출력되도록 프로그램 생성

# 1
# 22
# 333
# 4444
# 55555

for i in range(1, 6):
    print(f"{str(i) * i}")
```

```python
# 2. 아래와 같이 출력되도록 프로그램 생성

# *****
#  ***
#   *

for i in range(1, 6):
    blank = " " * (i -1)
    star = "*" * (7 - (i*2))

    print(blank + star)
```
