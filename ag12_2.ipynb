```python
n = int(input("숫자 개수 : "))

numbers = []

for i in range(n):
    numbers.append(i)

print(numbers)
```

```python
n = int(input("크기 : "))

matrix = []

for i in range(n):
    row = []
    for j in range(n):
        row.append(0)
    matrix.append(row)

print(matrix)
```
