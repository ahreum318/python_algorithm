```python
# re.match(패턴, 문자열)

import re
text = "Python"
result = re.match(r"Py", text)
print(result)
print(result.group())
```

```python
import re

text = "Python"
result = re.match(r"th", text)
print(result) # None match()는 문자열의 시작에서만 검사

text = "Python"
print(re.search(r"th", text)) # re.search() 문자열 전체를 검색한다.
```

```python
# re.findall() 모든 검색 결과를 리스트로 반환한다.

import re

text = "apple banana orange"
result = re.findall(r"a", text)
print(result) # ['a', 'a', 'a', 'a', 'a']
```

```python
import re

text = "A10 B20 C30"
numbers = re.findall(r"\d+", text) # \d 숫자한문자, \d+ 연속된 숫자 한문자이상
print(numbers)
```

```python
# re.sub(패턴, 변경문자, 문자열)

# 숫자 문자 제거
import re

text = "A10 B20 C30"
result = re.sub(r"\d", "", text)
print(result)
```

```python
# 실습1) 상품명에서 가격만 출력

products = [
    "사과 3500원",
    "배 4200원",
    "수박 18000원"
]

price_list = []

for product in products:
    price = re.search(r"\d+", product)
    #print(price.group())
    price_list.append(price.group())

print(price_list)
```

```python
# 실습2) 여러 공백을 하나의 공백으로 치환

import re

text = "Python   Data   Analysis"
result = re.sub(r"\s+", " ", text) # \s+ 1개이상의 공백 문자
print(result)
```

```python
# 실습3) 이메일 검증

import re

emails = [
    "python@test.com",
    "kim123@gmail.com",
    "abc@company.co.kr",
    "hello",
    "@gmail.com"
]

pattern = r"^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$"

for email in emails:
    if re.fullmatch(pattern, email):
        print(email, "정상")
    else:
        print(email, "오류")
```

```python
# 실습4) URL 검증

import re

urls = [
    "https://www.google.com",
    "http://naver.com",
    "https://python.org/docs",
    "google.com",
    "https://cafe.naver.com/f-e/cafes/25331457/articles/19125?menuid=304&referrerAllArticles=false"
]

pattern = r"^https?://([\w-]+\.)+[\w-]+(/[^\s]*)?$"
# ^https? - http나 https로 시작해야됨
# "://" - "http://" or "https://" 시작
# [\w-]+ - ["\w"이거나 "-"]
# ([\w-]+\.) - [\w-] 뒤에 무조건 점이 와야 한다 / httpL//abc.
# (/[^\s]*)?$ - com

for url in urls:
    if re.fullmatch(pattern, url):
        print(url, "정상")
    else:
        print(url, "오류")
```

```python
# 실습5) 주민등록변호 형식 검사(외국사람 제외)

import re

nums = [
    "901010-1234567",
    "000101-4234567",
    "9010101234567",
    "90101-1234567"
]

pattern = r"^\d{6}-[1-4]\d{6}$"

for n in nums:
    if re.fullmatch(pattern, n):
        print(n, "형식 정상")
    else:
        print(n, "오류")
```
