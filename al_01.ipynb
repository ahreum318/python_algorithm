```python
# 1. 팀장 번호 구하기 (산술 연산자)

# 1번부터 100번까지 학생이 있다.
# 1~10번은 1팀 
# 11~20번은 2팀 
# 21~30번은 3팀 
# ... 
# 각 팀은 10명씩이며 첫 번째 학생이 팀장이다.
# 학생 번호를 입력받아 해당 학생의 팀장 번호를 출력하는 프로그램을 작성하시오.

#실행 예
# 학생 번호 : 8
# 팀장 번호 : 1

# 학생 번호 : 15
# 팀장 번호 : 11

num = int(input("학생 번호: "))
# (num-1) // 10
# 1~10 → 0
leader_num = ((num-1) // 10) + 1

# 11~20 → 10
leader_num = ((num-1) // 10 * 10) + 1


print(leader_num)
```

```python
# 2. 홀수와 짝수 판별하기 (if~else)
# 정수를 입력받아 홀수인지 짝수인지 출력하는 프로그램을 작성하시오.

try:
    num = int(input("숫자를 입력하세요"))

    if num % 2 == 0:
        print("짝수")
    else:
        print("홀수")

except ValueError as e: 
    print("숫자만 입력하세요")
    #print(e)
```

```python
# 3. 양수, 음수, 0 판별하기 (if~elif~else)
# 정수를 입력받아 양수, 음수 또는 0을 출력하는 프로그램을 작성하시오.

try:
    num = int(input("정수를 입력하세요"))

    if num > 0:
        print("양수")
    elif num < 0:
        print("음수")
    else:
        print("0")

except ValueError as e:
    print("숫자만 입력하세요")
```

```python
# 4. 20대 출입 가능 여부
# 두 사람의 나이를 입력받는다.
# 두 사람 모두 20대(20~29세)이면 출입 가능을 출력하고,
# 그 외에는 출입 불가를 출력하는 프로그램을 작성하시오

try:
    age1 = int(input("첫 번째 사람의 나이: "))
    age2 = int(input("두 번째 사람의 나이: "))

    if (20 <= age1 <= 29) and (20 <= age2 <= 29):
        print("출입 가능")
    else:
        print("출입 불가")

except ValueError as e:
    print("나이를 입력하세요")
```

```python
# 6. 1~10 사이 홀수의 합 구하기
# 1부터 10까지의 홀수의 합을 출력하는 프로그램을 작성하시오.

# 01
total = 0

for i in range(1, 11):
    if i % 2 == 1:
        total += i

print("1~10 사이 홀수의 합:", total)


# 02
total = 0
for i in range(1, 11, 2):
    total += i
print("1~10 사이 홀수의 합:", total)
```

```python
#7. 윤년 구하기
# 2014년부터 2500년까지의 윤년을 모두 출력하는 프로그램을 작성하시오.

# 윤년의 조건
# 400으로 나누어 떨어지는 해 
# 또는 4로 나누어 떨어지지만 100으로는 나누어 떨어지지 않는 해

print("--- 윤년 리스트 ---")

for year in range(2014, 2501):
    if (year % 400 == 0) or ((year % 4 == 0) and (year % 100 != 0)):
        print(year, end=" ")
```

```python
# 8. 소수 판별하기
# 정수를 입력받아 소수인지 아닌지 판별하는 프로그램을 작성하시오.
# (소수 = 1보다 큰 자연수 중 1과 자기 자신만으로 나누어지는 수)

# = 1과 자기 자신 이외에는 나머지가 무조건 있어야 함

try:
    num = int(input("정수를 입력하세요: "))

    if num <= 1:
        print(f"{num}은 소수가 아닙니다.")
    else:
        for i in range(2, num):
            if num % i == 0:
                print(f"{num}은 소수가 아닙니다.")
                break
        else:
            print(f"{num}은 소수입니다.")

except ValueError as e:
    print("숫자를 입력하세요")
```

```python
# 9. 3~100 사이의 소수 개수 구하기
# 3부터 100까지의 소수의 개수를 출력하는 프로그램을 작성하시오.

total = 0

for num in range(3, 101):
    for i in range(2, num):
        if num % i == 0:
            break
    else:
        total +=1

print("3~100 사이의 소수 개수:",total)
print(f"3~100 사이의 소수 개수: {total}")
```

```python
# 10. 별 출력하기
# 다음과 같이 출력하는 프로그램을 작성하시오.

for i in range(1, 6):
    print("*" * i)
```

```python
# 번외 10-1. 오른쪽 정렬 삼각형
# 첫번째 줄: 공백 4/별 1
# 두번째 줄: 공백 3/별 2
# 세번째 줄: 공백 2/별 3
# 이렇게 공백은 -1 / 별 +1 규칙 있음

for i in range(1, 6):
    print(" " * (5 - i) + "*" * i)

for i in range(1, 6):
    blank = " " * (5 -i)
    star = "*" * i

    print(blank + star)
```

```python
# 번외 10-2. 왼쪽 정렬 역삼각형
# 별이 5, 4, 3, 2, 1 / -1로 줄어들며 출력되면 됨

for i in range(1, 6):
    print("*" * (6-i))
```

```python
# 번외 10-3. 오른쪽 정렬 역삼각형
# 첫번째 줄: 공백 1/별 4
# 두번째 줄: 공백 2/별 3
# 세번째 줄: 공백 3/별 2
# 이렇게 공백은 +1 / 별 -1 규칙 있음

for i in range(1, 6):
    print(" " * (i - 1) + "*" * (6 - i))

for i in range(1, 6):
    blank = " " * (i -1)
    star = "*" * (6 - i)

    print(blank + star)
```

```python
#11. X 모양 별 출력
#다음과 같이 출력하는 프로그램을 작성하시오.

# 중간 공백 : 2씩 줄어들다 0이 되는 순간 1번 반복하고 다시 +2
# 사이드 공백 : +1씩 늘어남
# 별 : 한 줄당 총 2개, 자리가 +1 / -1 반복

for i in range(0, 5):
    print(" " * i + "*" + " " * (8 - 2*i) + "*")

for i in range(4, -1, -1):
    print(" " * i + "*" + " " * (8 - 2*i) + "*")
```

```python
# 11-1. AI라면..

for y in range(10):  # 세로 0~9줄
    for x in range(10):  # 가로 0~9칸
        # 규칙 1: x와 y가 같거나 / 규칙 2: x와 y를 더해서 9가 되면 별 찍기
        if x == y or x + y == 9:
            print("*", end="")   # 줄바꿈 안 하고 옆으로 붙여 쓰기
        else:
            print(" ", end="")  # 별이 아니면 공백 찍기
    print()  # 한 줄 다 찍으면 다음 줄로 넘어가기
```

```python
#12. 가로 구구단 출력
#2단부터 9단까지를 가로 방향으로 출력하는 프로그램을 작성하시오.
#출력 포맷 답 부분은 2자리 숫자로 출력 ex) "2*1=2" -> "2*1= 2"

for j in range(2, 10):
    for i in range(1, 10):
        result = j * i

        print (f" {j}*{i}={result:2d}", end=" ｜")
    print()
```

```python
# 13. 세로 구구단 출력
# 2단부터 9단까지를 세로 방향으로 출력하는 프로그램을 작성하시오.
# 출력 포맷 답 부분은 2자리 숫자로 출력 ex) "2*1=2" -> "2*1= 2"

for i in range(1, 10):
    for j in range(2, 10):
        result = i * j

        print (f" {j}*{i}={result:2d}", end=" ｜")
    print()
```
