```python
from collections import deque

dq = deque()
dq.append(10)
dq.append(20)
dq.append(30) #0번째 위치에 삽입

print(dq)

x = dq.pop() #스택처럼...

print(x)
print(dq)

print(dq.popleft()) #queue처럼...
print(dq)
```

```python
# 스택이 필요한 경우 > 효율성 때문에 Deque 사용 X (list 사용)
# 큐가 필요한 경우 > 효율성 때문에 Deque 사용 O (시계방향)
# 덱 필요한 경우 > Deque 사용
# 자료구조의 회전 > 시계/반시계 방향에 따라 양방향 삽입/삭제 가능한 Deque 사용
```

```python
# 회전 roate

dq = deque([1, 2, 3, 4, 5])
dq.rotate(1) # 우측으로 한칸 이동(회전)
print(dq) # deque([5, 1, 2, 3, 4])

dq1 = deque([1, 2, 3, 4, 5])
x = dq1.pop()
dq1.appendleft(x)
print(dq1) # deque([5, 1, 2, 3, 4])
```

```python
# 기본 모듈 중 덱을 사용하는 이유는 자료구조 때문
# 편리성 메서드를 많이 제공함 (큐, 스텍, 회전, 최대개수, ...)
```

```python
#​ - 최근 방분 기록 n개 저장
history = deque(maxlen=5) # maxlen=n n개만 저장

history.append("google")
history.append("naver")
history.append("youtube")
history.append("github")
history.append("chatgpt")

print(history)

history.append("python") #제일 앞에 있는 구글이 빠짐

print(history)
```

```python
#- 카드 게임(원카드, 우노, ..) 시뮬레이션
cards = deque([1, 2, 3, 4, 5]) #처음 받은 카드 5장

while len(cards) > 1:
    print("버림 :", cards.popleft()) #1번 카드 버림
    cards.rotate(-1)
    #cards.append(cards.popleft())
    print("남은 카드 :", cards)

print("마지막 카드 :", cards[0])
```
