```python
# 실습 4. 객체 정보 출력: 객체가 자신의 정보를 출력하도록 한다.
# 요구사항: print_info() 메서드를 작성한다.
# 반복문으로 모든 상품 정보를 출력한다.

class Product:
    def __init__(self, name, price, stock):
        self.name = name
        self.price = price
        self.stock = stock

    def total_price(self, count):
        return self.price * count
    
    def print_info(self):
        print(f"상품명 : {self.name}")
        print(f"가격 : {self.price}원")
        print(f"재고 : {self.stock}개")
        print("-" * 20)

products = [
    Product("키보드", 50000, 10),
    Product("마우스", 30000, 20),
    Product("모니터", 250000, 5),
    Product("노트북", 1200000, 3)
]

for p in products:
    p.print_info()
```

```python
# 실습 5. 객체의 상태 변경: 객체의 속성은 계속 변경될 수 있음을 이해한다.
# 다음 메서드를 작성한다. : 입고(stock_in) / 출고(stock_out) / 가격변경(change_price)

class Product:
    def __init__(self, name, price, stock):
        self.name = name
        self.price = price
        self.stock = stock

    def total_price(self, count):
        return self.price * count
    
    def print_info(self):
        print(f"상품명 : {self.name}")
        print(f"가격 : {self.price}원")
        print(f"재고 : {self.stock}개")
        print("-" * 20)

    def stock_in(self, count):
        self.stock += count

    def stock_out(self, count):
        if self.stock >= count:
            self.stock -= count
        else:
            print(f"❌ {self.name} 재고가 부족하여 출고할 수 없습니다! (현재 재고: {self.stock}개)")

    def change_price(self, new_price):
        if new_price >= 0:
            self.price = new_price
        else:
            print("❌ 가격은 0원보다 작을 수 없습니다!")

products = [
    Product("키보드", 50000, 10),
    Product("마우스", 30000, 20),
    Product("모니터", 250000, 5),
    Product("노트북", 1200000, 3)
]

products[0].stock_in(5)
products[1].stock_out(3)
products[2].change_price(270000)

for p in products:
    p.print_info()
```

```python
# 실습 6. 메서드를 이용한 기능 구현: 객체가 자신의 데이터를 스스로 처리하도록 만든다.
"""
요구사항

- 다음 메서드를 추가한다.
  order(count)
  total_price(count)
  is_sold_out()

- 기능
  주문하면 재고 감소 
  주문 금액 계산 
  품절 여부 반환 

출력
- 키보드 2개 주문
- 결제금액 : 100000원
- 남은 재고 : 8개

추가 문제
- 주문 수량이 재고보다 많으면 재고 부족을 출력한다.
"""

class Product:
    def __init__(self, name, price, stock):
        self.name = name
        self.price = price
        self.stock = stock

    def total_price(self, count):
        return self.price * count
        
    def print_info(self):
        print(self.name)
        print(self.price)
        print(self.stock)

    def is_sold_out(self):
        return self.stock <= 0
    
    def order(self, count):
        if count > self.stock:
            print("재고 부족")
        else:
            money = self.total_price(count)
            self.stock -= count

            print(f"- {self.name} {count}개 주문")
            print(f"- 결제금액 : {money}원")
            print(f"- 남은 재고 : {self.stock}개")

products = [
    Product("키보드", 50000, 10),
    Product("마우스", 30000, 20),
    Product("모니터", 250000, 5),
    Product("노트북", 1200000, 3)
]

products[0].order(2)

```

```python
# 실습 7. 딕셔너리 기반 프로그램을 클래스로 리팩터링
# Step 1. 기존 코드
products = [
    {"name":"키보드","price":50000,"stock":10},
    {"name":"마우스","price":30000,"stock":20},
    {"name":"모니터","price":250000,"stock":5},
]

for p in products:
    print(p["name"])
    print(p["price"])
    print(p["stock"])
# 딕셔너리를 사용하면 주문 기능, 재고 감소,  품절 여부 검사, .. 
# 이런 기능들은 데이터와 별도로 함수로 구현 or 데이터 아래 절차적으로 구현 되어야 한다.

# Step 2. 클래스로 변경: 딕셔너리를 모두 Product 객체로 변경한다.
# Step 3. 기능 구현: 절차적으로 구현된 재고 감소, 가격 계산, 품절 확인 등의 기능을
"""
상품 등록
전체 상품 조회
상품 검색
주문하기
입고하기
출고하기
가격 변경
할인 적용(10%, 20%)
품절 상품 조회
가장 비싼 상품 찾기
재고가 가장 많은 상품 찾기
전체 상품 재고 수량 합계
전체 재고 금액(가격 × 재고) 계산
"""
# 의 메서드 형태로 변경한다.
```

```python
class Product:
    def __init__(self, name, price, stock):
        self.name = name
        self.price = price
        self.stock = stock

    # 1. 상품 정보 출력 (실습 4 완료)
    def print_info(self):
        print(f"상품명 : {self.name} | 가격 : {self.price}원 | 재고 : {self.stock}개")

    # 2. 주문 금액 계산 (실습 6 완료)
    def total_price(self, count):
        return self.price * count

    # 3. 품절 여부 반환 (실습 6 완료)
    def is_sold_out(self):
        return self.stock <= 0

    # 4. 입고하기 (실습 5 완료)
    def stock_in(self, count):
        self.stock += count

    # 5. 출고하기 (실습 5 완료)
    def stock_out(self, count):
        if self.stock >= count:
            self.stock -= count
        else:
            print("재고 부족")

    # 6. 가격 변경 (실습 5 완료)
    def change_price(self, new_price):
        if new_price >= 0:
            self.price = new_price

    # 7. 주문하기 (실습 6 완료)
    def order(self, count):
        if count > self.stock:
            print("재고 부족")
        else:
            money = self.total_price(count)
            self.stock -= count
            print(f"🛒 {self.name} {count}개 주문 완료 (결제: {money}원)")

    def apply_discount(self, percent):
        self.price = int(self.price * (1-percent / 100))

    def get_total_asset(self):
        return self.price * self.stock
    
products = [
    Product("키보드", 50000, 10),
    Product("마우스", 30000, 20),
    Product("모니터", 250000, 5)
]

print("=== 1. 전체 상품 조회 ===")
for p in products:
    p.print_info()

print("\n=== 2. 상품 검색 (예: 마우스) ===")
for p in products:
    if p.name == "마우스":
        p.print_info()

print("\n=== 3. 품절 상품 조회 ===")
for p in products:
    if p.is_sold_out():
        print(f"⚠ 품절: {p.name}")

# 계산을 위해 변수들을 미리 0개, 혹은 첫 번째 상품으로 세팅해둡니다!
most_expensive = products[0]
most_stock = products[0]
total_stock_count = 0
total_asset_money = 0

for p in products:
    # 4. 가장 비싼 상품 찾기 복사
    if p.price > most_expensive.price:
        most_expensive = p
        
    # 5. 재고가 가장 많은 상품 찾기
    if p.stock > most_stock.stock:
        most_stock = p
        
    # 6. 전체 상품 재고 수량 합계 누적
    total_stock_count += p.stock
    
    # 7. 전체 재고 금액 합계 누적
    total_asset_money += p.get_total_asset()

print(f"💰 가장 비싼 상품 : {most_expensive.name} ({most_expensive.price}원)")
print(f"📦 재고가 가장 많은 상품 : {most_stock.name} ({most_stock.stock}개)")
print(f"📊 전체 상품 재고 수량 합계 : {total_stock_count}개")
print(f"💵 전체 재고 금액 총합 : {total_asset_money}원")

print("\n=== 8. 키보드에 10% 할인 적용하기 ===")
products[0].apply_discount(10)
products[0].print_info()
```
