```python
# 1) Raw String
# 백슬래시(\) 문자를 이스케이프 문자로 해석하지 않고 그대로 저장하는 문자열 표현

print("Hello\nworld\\")
print(r"Hello\nworld\\")
```

```python
# 2) 문자클래스 []

import re

text = "apple banana cat"
result = re.findall(r"[abc]", text)
    # "apple banana cat"를 0부터 끝까지 "a", "b", "c" 중 일치하면 반환되는 리스트에 추가
    # 답이 없어도 리스트 반환함 []
print(result)
```

```python
text = "apple banana cat"
result = re.findall(r"[at]", text)
print(result)
if result:
    print("a나 t가 있다")
else:
    print("a나 t가 없다")
```

```python
# 3) 대괄호[] 안의 "-"는 범위를 나타낸다.
# 문자0부터 문자9까지 [0123456789]를 줄여서 [0-9]로 표현
# [a-z], [A-Z], ...

text = "A12B34"
print(re.findall(r"[0-9]", text))
```

```python
import re

text = "Python JAVA AI 123456789"
print(re.findall(r"[a-zA-Z0-9]", text))
print(re.findall(r"[^0-9]", text)) #대괄호 안의 ^ = not, 숫자 제외 전부 반환함
```

```python
import re

text = "a7bc123"
result = re.findall(r"[^0-9]", text) # 0~9문자가 아니면 반환되는 리스트에 추가
print(result)
```

```python
#소대문자 포함, 특수문자 포함, 숫자문자 포함
#비밀번호 만들기

import re
password = input("비밀번호 : ")

result = []

if re.search(r"[ㄱ-ㅎㅏ-ㅣ가-힣]", password):
    result.append("한글을 사용할 수 없습니다.")

if not re.search(r"[a-z]", password):
    result.append("소문자가 없습니다.")

if not re.search(r"[A-Z]", password):
    result.append("대문자가 없습니다.")

if not re.search(r"[0-9]", password):
    result.append("숫자가 없습니다.")

if not re.search(r"[^a-zA-Z0-9]", password):
    result.append("특수문자가 없습니다.")

if len(result) == 0:
    result.append("사용 가능한 비밀번호입니다.")

for r in result:
    print(r)
```

```python
import re
password = input("비밀번호 : ")

# 💡 실전용 단 한 줄짜리 마법의 정규식 패턴
# (영어 소문자, 대문자, 숫자, 특수문자가 각각 최소 1개 이상 포함되어야 함)
pattern = r"^(?=.*[a-z])(?=.*[A-Z])(?=.*[0-9])(?=.*[^a-zA-Z0-9ㄱ-ㅎㅏ-ㅣ가-힣])[^\sㄱ-ㅎㅏ-ㅣ가-힣]+$"

if re.match(pattern, password):
    print("사용 가능한 비밀번호입니다.")
else:
    print("비밀번호 규칙에 맞지 않습니다. (영문 대/소문자, 숫자, 특수문자 포함, 한글 제외)")
```
