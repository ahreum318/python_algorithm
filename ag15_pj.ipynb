```python
# 1. 문자열에서 숫자를 모두 찾아 출력하시오.

import re

text = "Python123AI456"
result = re.findall(r"\d+", text)

print(result)
```

```python
# 2. 문자열에서 소문자만 모두 출력하시오.

text = "PythonAI"
result = re.findall(r"[a-z]", text)

print(result)
```

```python
# 3. 문자열에 숫자가 포함되어 있는지 검사하시오.

text = "Python123"

if re.search(r"\d", text):
    print("숫자가 있습니다")
else:
    print("숫자가 없습니다")
```

```python
# 4. 문자열에서 전화번호를 찾아 출력하시오.
text = "문의 : 010-1234-5678"

pattern = r"010-\d{4}-\d{4}"
result = re.search(pattern, text)

print(result.group())
```

```python
# 5. 문자열에서 이메일 찾아 출력하시오.

text = "문의 : python@test.com"

pattern = r"\S+@\S+\.\S+"
#pattern = r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}"
result = re.search(pattern, text)

print(result.group())
```

```python
# 6. 문자열에서 숫자를 모두 *로 변경하시오.

text = "A10 B20 C30"

result = re.sub(r"\d", "*", text)
print(result)
```

```python
# 7. 문자열에서 HTML태그("<"로 시작하고 ">"로 끝나는 문자)를 제거하시오.

html = "<h1>Python</h1>"

pattern = r"<.*?>" #?를 붙여서 가장 가까운 >까지만 묶으라고 지시
result = re.sub(pattern, "", html)

print(result)
```

```python
# 8. 문자열에서 날짜를 찾아 출력하시오.

text = """
오늘은 2026-07-07입니다.
시험은 2026-08-01입니다.
"""

pattern = r"\d{4}-\d{2}-\d{2}"
result = re.findall(pattern, text)

print(result)
```

```python
# 9. 아이디는 영문 소문자와 숫자로만 이루어진 5~10글자이다.
# 다음 아이디중 조건에 맞는 것만 출력하시오.

ids = [
    "python",
    "abc123",
    "ABC123",
    "py",
    "python12345",
    "123python"
]

pattern = r"[a-z0-9]{5,10}" #{5, 10} 5 이상 10 이하
for id in ids:
    if re.fullmatch(pattern, id):
        print(id)
```
