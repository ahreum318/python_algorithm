```python
# 실습 과제 1. 누적합(Prefix Sum)
# 1. 한 학생의 7일 동안의 걸음 수가 리스트에 저장되어 있다.

steps = [5231, 6120, 4890, 7005, 6500, 7200, 6800]
# 다음과 같이 매일 누적 걸음 수를 계산하는 프로그램을 작성하시오.

prefix = [0]

for s in steps:
    prefix.append(prefix[-1] + s)

print("누적 걸음 수:", prefix)
```

```python
# 실습 과제 2. 누적합(Prefix Sum) - 1번 응용
# 2. 한 학생의 7일 동안의 걸음 수가 리스트에 저장되어 있다.

steps = [
    {"2026-01-01": 5231},
    {"2026-01-02": 6210},
    {"2026-01-03": 4890},
    {"2026-01-04": 7005},
    {"2026-01-05": 6500},
    {"2026-01-06": 7200},
    {"2026-01-07": 6800}
]

"""
2026-01-01 누적 : 5231
2026-01-02 누적 : 11441
2026-01-03 누적 : 16331
2026-01-04 누적 : 23336
2026-01-05 누적 : 29836
2026-01-06 누적 : 37036
2026-01-07 누적 : 43836
""" #이렇게 출력

total = 0

for d in steps:
    for date, val in d.items():
        total += val
        print(f"{date} 누적 : {total}")


for d in steps:
    date = list(d.keys())[0]
    val = d[date]

    total += val
    print(f"{date} 누적 : {total}")        
```

```python
# 실습 과제 3. 슬라이딩 윈도우
# 3. 한 카페의 시간대별 매출이 다음과 같다.

sales = [12, 15, 18, 20, 17, 25, 22, 19]
# 연속된 3시간의 매출 합을 모두 구하여 출력하고, 가장 큰 합을 출력하는 프로그램을 작성하시오.

"""
실행결과
12 + 15 + 18 = 45
15 + 18 + 20 = 53
18 + 20 + 17 = 55
20 + 17 + 25 = 62
17 + 25 + 22 = 64
25 + 22 + 19 = 66

최대 매출 합 : 66
"""

p = 3 #window size
window = 0
for i in range(p):
    window += sales[i]

# 첫 번째 구간(12 + 15 + 18 = 45) 출력하기
print(f"{sales[0]} + {sales[1]} + {sales[2]} = {window}")

max_sum = window # 최대 매출 합을 저장할 변수
for i in range(p, len(sales)):
    window = window + sales[i] - sales[i-p]
    print(f"{sales[i-2]} + {sales[i-1]} + {sales[i]} = {window}")
    if window > max_sum:
        max_sum = window

print()
print(f"최대 매출 합 : {max_sum}")

```

```python
# 실습 과제 4. 슬라이딩 윈도우 - 3번 응용
# 4. 한 카페의 시간대별 매출이 다음과 같다.

sales = [
    {"9시": 12},
    {"10시": 15},
    {"11시": 18},
    {"12시": 20},
    {"13시": 17},
    {"14시": 25},
    {"15시": 22},
    {"16시": 19},
    {"17시": 24},
    {"18시": 21},
    {"19시": 18},
    {"20시": 16},
    {"21시": 14},
    {"22시": 11}
]

# 연속된 3시간의 매출 합을 모두 구하여 출력하고, 가장 큰 매출 합과 시간대를 출력하는 프로그램을 작성하시오.

"""
실행 결과

9시 ~ 11시 : 45
10시 ~ 12시 : 53
11시 ~ 13시 : 55
12시 ~ 14시 : 62
13시 ~ 15시 : 64
14시 ~ 16시 : 66
15시 ~ 17시 : 65
16시 ~ 18시 : 64
17시 ~ 19시 : 63
18시 ~ 20시 : 55
19시 ~ 21시 : 48
20시 ~ 22시 : 41

최대 매출 합 : 66
시간대 : 14시 ~ 16시
"""
```

```python
sales = [
    {"9시": 12},
    {"10시": 15},
    {"11시": 18},
    {"12시": 20},
    {"13시": 17},
    {"14시": 25},
    {"15시": 22},
    {"16시": 19},
    {"17시": 24},
    {"18시": 21},
    {"19시": 18},
    {"20시": 16},
    {"21시": 14},
    {"22시": 11}
]


time_list = []
price_list = []

for d in sales:
    for time, price in d.items():
        time_list.append(time)
        price_list.append(price)

p = 3
window = 0
for i in range(p):
    window += price_list[i]

print(f"{time_list[0]} ~ {time_list[2]} : {window}")

max_sum = window
max_time = f"{time_list[0]} ~ {time_list[2]}"

for i in range(p, len(price_list)):
    window = window + price_list[i] - price_list[i-p]

    start_time = time_list[i-2]
    end_time = time_list[i]
    print(f"{start_time} ~ {end_time} : {window}")

    if window > max_sum:
        max_sum = window
        max_time = f"{start_time} ~ {end_time}"

print()
print(f"최대 매출 합 : {max_sum}")
print(f"시간대 : {max_time}")
```

```python
# 번외: 제미나이 문제

stocks = [
    {"월요일": 50000},
    {"화요일": 52000},
    {"수요일": 49000},
    {"목요일": 55000},
    {"금요일": 54000}
]

"""
월요일 ~ 화요일 : 102000
화요일 ~ 수요일 : 101000
수요일 ~ 목요일 : 104000
목요일 ~ 금요일 : 109000

최대 주가 합 : 109000
시간대 : 목요일 ~ 금요일
"""

date_list = []
price_list = []

for d in stocks:
    for date, price in d.items():
        date_list.append(date)
        price_list.append(price)

p = 2
window = 0
for i in range(p):
    window += price_list[i]

print(f"{date_list[0]} ~ {date_list[1]} : {window}")

max_sum = window
max_date = f"{date_list[0]} ~ {date_list[1]}"

for i in range(p, len(price_list)):
    window = window + price_list[i] - price_list[i-p]

    start_date = date_list[i-1]
    end_date = date_list[i]
    print(f"{start_date} ~ {end_date} : {window}")

    if window > max_sum:
        max_sum = window
        max_date = f"{start_date} ~ {end_date}"

print()
print(f"최대 주가 합 : {max_sum}")
print(f"시간대 : {max_date}")
    
```
