```python
class Student:
    def __init__(self, name):
        self.name = name
    def print_name(self):
        print(self.name)
    def greet(self):
        print(f"{self.name}님 hello")
```

```python
# 상속하면
class HighStudent(Student):
    def study(self):
        print("공부한다.")
```

```python
# 사용
s = HighStudent("홍길동")
s.print_name()
s.study()
s.greet()
```

```python
# 1) 딕셔너리 함수
def average(student):
    return (student["kor"] + student["eng"]) / 2


student = {
    "name": "홍길동",
    "kor": 90,
    "eng": 85,
    "average": average
}

print(student["average"](student))


student2 = {
    "name": "김철수",
    "kor": 70,
    "eng": 75,
    "average": average
}

print(student2["average"](student2))
```

```python
# 2) 클래스 메서드

class Student:

    def __init__(self, name, kor, eng):
        self.name = name
        self.kor = kor
        self.eng = eng
        self.math = 0

    def average(self):
        return (self.kor + self.eng + self.math) / 3


s = Student("홍길동", 90, 85)
s.math = 88
s2 = Student("김철수", 50, 60)
s2.math = 70

print(s.average())
print(s2.average())
```

```python
# 2) 클래스로 변경
# 평균 계산 방법이 바뀌면 클래스 내부의 메서드 코드만 변경하면 된다.

class Student:

    def __init__(self, name, kor, eng):
        self.name = name
        self.kor = kor
        self.eng = eng

    def total(self):
        return self.kor + self.eng

    def average(self):
        return self.total() / 2


s1 = Student("홍길동", 90, 85)
print(s1.average())

s2 = Student("김철수", 80, 88)
print(s2.average())
```
