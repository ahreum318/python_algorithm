```python
nums = [1, 3, 5, 7, 9, 10, 20, 10]

# 1번째부터 8번째
# 인덱스 0부터 7까지

t1 = 0
for i in range(8): #0~7
    t1 = t1 + nums[i]

print(t1)


# 1번째부터 3번째까지
t2 = 0
for i in range(3): #0~2
    t2 = t2 + nums[i]
print(t2)

```

```python
# 누적합 배열을 만들어 두기

nums = [1, 3, 5, 7, 9, 10, 20, 10]

prefix = [0] #초기값 0 필요

for n in nums:
    # 0 + 1
    # prefix[1] + 3
    # prefix[2] + 5
    # prefix[3] + 7
    # ...
    prefix.append(prefix[-1] + n)

print(nums)
print(prefix)
```

```python
# 처음부터 끝까지의 합 (1부터 8번째) (0, 7)
prefix[8]
```

```python
#처음부터 7까지의 합 (1부터 7번째) (0, 6)
prefix[7]
```

```python
#원본의 5번째에서 7번째까지 (4, 6)
s = 5
e = 7

prefix[7] - prefix[s-1]
```
