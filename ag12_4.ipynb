```python
queue = ["A", "B", "C"]

pop_index = 0
while pop_index < len(queue):
    print(queue[pop_index])
    print(queue[pop_index+1:])
    pop_index += 1
print(queue)
```

```python
# 프린트 출력 큐 예제

queue = []

queue.append("문서1")
queue.append("문서2")
queue.append("문서3")

while len(queue) > 0:
    print("출력 :", queue.pop(0))
```

```python
# 4. Stack vs Queue

stack = []

stack.append("A")
stack.append("B")
stack.append("C")

while len(stack) > 0:
    print(stack.pop(), end=" ")
```

```python
queue = []

queue.append("A")
queue.append("B")
queue.append("C")

while len(queue) > 0:
    print(queue.pop(0), end=" ")
```

```python
queue = ["A", "B", "C"]

pop_index = 0

while pop_index < len(queue):
    print(queue[pop_index]) #["A", "B", "C"]
    print(queue[pop_index+1:])
    queue = queue[pop_index+1:]
```

```python
from collections import deque

queue = deque()

queue.append("A")
queue.append("B")
queue.append("C")

print(queue.popleft())
print(queue)
print(queue.pop())
print(queue)
```
