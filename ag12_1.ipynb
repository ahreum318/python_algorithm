```python
# 웹 브라우저 "뒤로가기" 구현

stack = []

stack.append("네이버")
stack.append("구글")
stack.append("ChatGPT")

print("현재 방문 기록 :", stack)
print("뒤로가기 :", stack.pop()) #뒤로가기 버튼 클릭이라면
print("현재 페이지 :", stack[-1]) #pop 연산 후 top 연산
print("뒤로가기 :", stack.pop())
print("현재 페이지 :", stack[-1])
```

```python
# 선형 탐색 - 반복 순환

numbers = [15, 8, 23, 7, 31, 18]
target = 18

count = 0

for i in range(len(numbers)):
    count += 1
    if numbers[i] == target:
        print("찾은 위치 :", i)
        break # 찾으면 남은 원소는 비교하지않고 break

print("비교 횟수 :", count)
```

```python
#최댓값 찾기

numbers = [15, 8, 23, 7, 31, 18]

max_num = numbers[0]
count = 0

for n in numbers:
    count += 1
    if n > max_num:
        max_num = n

print("최댓값 :", max_num)
print("비교 횟수 :", count)
```

```python
# ===========================
# 선형 탐색 (O(n))
# ===========================

# 실행 시간 측정 시작
start = time.time()

# 마지막 원소를 찾도록 설정(최악의 경우: 최대 실행 횟수 확인)
target = numbers[-1]
count = 0

# 처음부터 끝까지 하나씩 비교
for n in numbers:
    count += 1
    if n == target:
        break

# 실행 시간 측정 종료
end = time.time()

print("선형 탐색 :", end - start, "count:", count)
```

```python
# ===========================
# 선형 탐색 (O(n))
# ===========================

# 실행 시간 측정 시작
start = time.time()

# 마지막 원소를 찾도록 설정(최악의 경우: 최대 실행 횟수 확인)
target = numbers[-1]
count = 0

# 처음부터 끝까지 하나씩 비교
for n in numbers:
    count += 1
    if n == target:
        break

# 실행 시간 측정 종료
end = time.time()

print("선형 탐색 :", end - start, "count:", count)
```

```python
# ===========================
# 버블 정렬 (O(n²))
# ===========================

# 새로운 무작위 데이터 생성
numbers = random.sample(range(10000), 7000)

# 실행 시간 측정 시작
start = time.time()

# 버블 정렬 수행
for j in range(len(numbers) - 1):
    for i in range(len(numbers) - 1 - j):
        if numbers[i] > numbers[i + 1]:
            numbers[i], numbers[i + 1] = numbers[i + 1], numbers[i]

# 실행 시간 측정 종료
end = time.time()

print("버블 정렬 :", end - start)
```
