```python
fruit_tree = {
    "question": ("color", "Red"),

    "yes": {
        "question": ("size", "Large"),
        "yes": "사과 (Apple)",
        "no": "체리 (Cherry)"
    },

    "no": {
        "question": ("color", "Purple"),
        "yes": "포도 (Grape)",
        "no": "모르는 과일"
    }
}
```

```python
#입력값 = 딕셔너리
fruit = {
    "color": "Red",
    "size": "Large" # Small
}

fruit_tree = {
    "question": ("color", "Red"),

    "yes": {
        "question": ("size", "Large"),
        "yes": "사과 (Apple)",
        "no": "체리 (Cherry)"
    },

    "no": {
        "question": ("color", "Purple"),
        "yes": "포도 (Grape)",
        "no": "모르는 과일"
    }
}

node = fruit_tree #node 변수의 초기값은 dict -> node의 변수값이 dict가 아닐 때까지 무한루프

# 트리를 탐색하는 반복문
while isinstance(node, dict): # node객체가 dict 이라면 True 아니면 False

    # dict이라면 "question"키값을 찾아 질문은 key, 정답은 value에 저장 
    key, value = node["question"] 
   
    # 입력데이터의 key값과 비교하여 정답이면 "yes"키값을 node에 대입, 아니면 "no"키값을 node에 대입
    # 이렇게 대입된 node값이 dict 타입이면 다시 반복문 진행... 
    # 이렇게 node에 dict타입이 아닌 타입(위 트리에선 문자열)이 올때 까지 반복문은 계속
    if fruit[key] == value:
        node = node["yes"]
    else:
        node = node["no"]

# 반복문을 빠져 나오면 마지막 node값을 출력
print(node)
```
